```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
decay = [0, 0.5, 0.75, 1, 1.5, 2]
acc_decay = [67.67, 83.31, 86.88, 88.92, 88.76, 88.90]
```

```python
dim = [1, 2, 4, 8, 16, 32, 64]
acc_dim = [88.71, 89.08, 88.77, 88.92, 88.76, 88.3, 88.52]
```

```python
# plt.style.use('seaborn-dark')
plt.rcParams.update(
    {
        "font.serif": ["Times"],
        "font.size": 16
    }
)
fig = plt.figure(figsize=(8, 4))
ax1 = fig.add_subplot(111)
ax2 = ax1.twiny()
ax1.set_xlabel('$t$, Decay speed=$1/pos^t$')
ax2.set_xlabel('Dimension of each scale')
line1 = ax1.plot(decay, acc_decay, label="Decay speed", color="red")
line2 = ax2.semilogx(dim, acc_dim, label="Dimension")
point1 = ax2.plot([8], [88.92], marker='o', markersize=5, color="purple", )
point2 = ax1.plot([0], [67.67], marker='o', markersize=5, color="purple", )
ax1.annotate('Decay: 1/pos, Dimension: 8', xy=(1.02, 88.5), xytext=(1, 86), arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))
ax1.annotate('No decay, Dimension: 8', xy=(0.02, 67.92), xytext=(0.2, 70), arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))
lines = line2 + line1
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc=4)
plt.savefig('decay_dim.pdf', bbox_inches="tight")
```
